```python
import random
import numpy as np

true_a = 0.04
true_b = 0.05

num_trials = 10000
```

```python
a_trials = [0,0]
b_trials = [0,0]

choice = []

for trial in range(num_trials):
    random_a = np.random.beta(a_trials[0] + 1, a_trials[1] + 1)
    random_b = np.random.beta(b_trials[0] + 1, b_trials[1] + 1)

    if (random_a > random_b):
        choice.append(0)
        if np.random.uniform() < true_a:
            a_trials[0] += 1
        else:
            a_trials[1] += 1
    else:
        choice.append(1)
        if np.random.uniform() < true_b:
            b_trials[0] += 1
        else:
            b_trials[1] += 1
        
```

```python
import matplotlib.pyplot as pyp

batch = 50

batched_choice = []

for index in range(int(num_trials / batch)):
    batched_choice.append(sum(choice[(index * batch):(index*batch + 50)]) / batch)
    

pyp.plot(batched_choice)
```
